```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Polynomial_Degree.csv")
```

```python
df.head()
```

```python
x = df["Polynomial Degree"]
y = df["Accuracy"]
```

```python
fig = plt.figure(figsize=(20,11))
plt.xlabel("Polynomial Degree", fontsize = 25)
plt.ylabel("Accuracy", fontsize = 25)
plt.title("Polynomial Degrees Accuracy Plot", fontsize = 25)
plt.plot(x, y, color = "g")
plt.tick_params(axis="x", colors='r', labelsize = 30)
plt.tick_params(axis="y", colors = "b" , labelsize = 20)

plt.savefig("Polynomial_Degrees_plot.png")
```
