```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv("Polynomial Regression Degree.csv")
```

```python
df.drop(columns = ["Unnamed: 2","Unnamed: 3", "Unnamed: 4", "Unnamed: 5", "Unnamed: 6"], inplace = True)
df.head()
```

```python
x = df["Degree of Polynomial Regression"]
y = df["Accuracy"]
```

```python
fig = plt.figure(figsize=(20,11))
plt.xlabel("Polynomial Degree", fontsize = 25)
plt.ylabel("Accuracy", fontsize = 25)
plt.title("Polynomial Regression Degrees Accuracy Plot", fontsize = 25)
plt.plot(x, y, color = "g")
plt.tick_params(axis="x", colors='r', labelsize = 30)
plt.tick_params(axis="y", colors = "b" , labelsize = 20)

plt.savefig("FPolynomial_Degrees_plot.png")
```
