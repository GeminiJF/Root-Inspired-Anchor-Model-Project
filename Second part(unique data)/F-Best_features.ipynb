```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Best Feature.csv")
df.drop(columns = ["Unnamed: 2","Unnamed: 3", "Unnamed: 4", "Unnamed: 5", "Unnamed: 6"], inplace = True)
df.dropna(inplace = True)
```

```python
df.head(20)
```

```python
x = df["Features"]
y = df["Accuracy"]
```

```python
fig = plt.figure(figsize=(20,11))
plt.xlabel(" Best Features", fontsize = 25)
plt.ylabel("Accuracy", fontsize = 25)
plt.title("Accuracy for Best Features", fontsize = 35)
plt.tick_params(axis="x", colors='r', labelsize = 17)
plt.tick_params(axis="y", colors = "g" , labelsize = 20)
plt.ylim(90, 98)
for a,b in zip(x, y): 
    plt.text(a, b, str(b), fontsize = 20, color = "orange")



plt.plot(x, y, linestyle = "-", marker = "o")
plt.savefig("FT_Accuracy_for_Unique_Best_Features.png")
```
