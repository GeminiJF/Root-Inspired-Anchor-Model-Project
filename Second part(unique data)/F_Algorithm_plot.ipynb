```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Predictive Model For f.xlsx")
df.to_csv("FPredictive Model For f.csv", index = False)
```

```python
df = pd.read_csv("FPredictive Model For f.csv")
```

```python
x = df["Algorithm"]
y = df["Accuracy"]
```

```python
fig = plt.figure(figsize=(20,15))

plt.xlabel("Predictive Models", fontsize = 25)
plt.ylabel("Accuracy", fontsize = 25)
plt.title("Accuracy for different Algorithms", fontsize = 35)
plt.tick_params(axis='x', labelsize = 25, rotation = 20)
plt.tick_params(axis="y", colors = "b" , labelsize = 20)
plt.ylim(70, 100)

for a,b in zip(x, y): 
    plt.text(a, b, str(b), fontsize = 20, color = "r")

plt.bar(x, y, color = "c", width = 0.5)
plt.savefig("F_Accuracy_different_algorithm4.png")
```
