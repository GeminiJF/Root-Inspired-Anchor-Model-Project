```python
df = pd.read_csv("neural_network_epoke.csv")
df2 = pd.read_csv("NN_plot.csv")
```

```python
df.head()
```

```python
df2.head()
```

```python
x = df["Epoke"]
x2 = df2["Number of Iteration"]
y1 = (df["Accuracy1"] +df["Accuracy2"])/2
y2 = df2["Accuracy(Tanh)"]
y3 = df2["Accuracy(selu)"]
```

```python
fig = plt.figure(figsize=(20,11))
plt.xlabel("number of iteration", fontsize = 25)
plt.ylabel("Accuracy", fontsize = 25)
plt.title("Neural Network", fontsize = 25)
plt.plot(x, y1, color = "g", label = "Activation: Sigmoid")
plt.plot(x2, y2, color = "orange", label = "Activation: Hyperbolic Tangent")
plt.plot(x2, y3, color = "blue", label = "Activation: Scale Exponential Linear Unit")
plt.legend(loc = 2, prop = {"size" : 20})
plt.ylim(0, 100)

plt.tick_params(axis="x", colors='r', labelsize = 30)
plt.tick_params(axis="y", colors = "b" , labelsize = 20)

plt.savefig("NN3a_plot.png")
```
