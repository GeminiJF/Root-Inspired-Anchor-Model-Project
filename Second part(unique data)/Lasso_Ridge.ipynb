```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Linear-change.csv")
```

```python
df.head()
```

```python
x = df["Linear Regression Models"]
y = df["Accuracy"]
```

```python
fig = plt.figure(figsize=(20,13))

plt.xlabel("Different Linear Models", fontsize = 25)
plt.ylabel("Accuracy", fontsize = 25)
plt.title("Linear vs Ridge vs Lasso", fontsize = 35)
plt.tick_params(axis='x', colors='orange', labelsize = 17)
plt.tick_params(axis="y", colors = "b" , labelsize = 20)
plt.ylim(75, 90)
for a,b in zip(x, y): 
    plt.text(a, b, str(b), fontsize =30, color = "r")

plt.bar(x, y, color = "c", width = 0.5)
plt.savefig("Ridge_Lasso_Linear.png")
```
