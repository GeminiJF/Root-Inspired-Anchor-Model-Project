```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

df = pd.read_csv("csv_project_dataset.csv")
```

```python
df.sort_values(by=["Pmax [N]"], inplace= True)
df.to_csv("sorted_by_Pmax_value_data.csv", index = False)
```

```python
df = pd.read_csv("sorted_by_Pmax_value_data.csv")
df_pmax = df["Pmax [N]"]
```

```python
maxi = df_pmax.max()
mini = df_pmax.min()
```

```python
one_third = (maxi - mini)/3
one_third
```

```python
two_third = (maxi - mini)*2/3
two_third
```

```python
fig = plt.figure(figsize=(15,8))
plt.axhline(y=one_third, color = "r", linestyle = "-", label = "First Trine(35.6[N])")
plt.plot(df_pmax, label = "Pullout force[N]")
plt.plot(one_third)
plt.axhline(y=two_third, color = "g", linestyle = "-", label = "Second Trine(71.3[N])")



plt.legend(loc = 2, prop = {"size" : 15})


plt.tick_params(axis='y', colors='orange', labelsize = 20)
plt.tick_params(axis='x', labelsize = 15)
plt.xlabel("Datapoints", fontsize = 20)
plt.ylabel("Pullout Force", fontsize = 20)
plt.title("Pullout Force Categorical Value - Trisector Plot", fontsize = 25)
plt.savefig("Categorical_Classification_Three_Division.png")
```
