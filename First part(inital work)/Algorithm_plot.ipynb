```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Algorithm_Accuracy.csv")
```

```python
df.head()
```

```python
x = df["Algorithm"]
y = df["Accuracy"]
```

```python
fig = plt.figure(figsize=(20,13))

plt.xlabel("Predictive Models", fontsize = 25)
plt.ylabel("Accuracy", fontsize = 25)
plt.title("Accuracy for different Algorithms", fontsize = 35)
plt.tick_params(axis='x', colors='red', labelsize = 17, rotation = -20)
plt.tick_params(axis="y", colors = "b" , labelsize = 20)

plt.bar(x, y, color = "g", width = 0.5)
plt.savefig("Accuracy_different_algorithms.png")
```
