```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

df = pd.read_csv("sorted_pmax_data_for_plot.csv")
```

```python
df_pmax = df["Pmax [N]"]
maxi = df_pmax.max()
mini = df_pmax.min()
```

```python
one_half = (maxi - mini)/2
one_half
```

```python
df_pmax
```

```python
fig = plt.figure(figsize=(15,8))
plt.axhline(y=one_half, color = "m", linestyle = "-", label = "Binary Value Boundary(53.4[N])")
plt.plot(df_pmax, label = "Pullout force[N]")



plt.legend(loc = 2, prop = {"size" : 15})


plt.tick_params(axis='y', colors='g', labelsize = 20)
plt.tick_params(axis='x', labelsize = 15)
plt.xlabel("Datapoints", fontsize = 20)
plt.ylabel("Pullout Force", fontsize = 20)
plt.title("Pullout Force Categorical Value - Binary Plot", fontsize = 25)
plt.savefig("Categorical_Classification_Binary.png")
```
